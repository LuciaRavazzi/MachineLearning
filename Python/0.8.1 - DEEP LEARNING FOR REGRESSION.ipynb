<h3><center> COMBINED CYCLE POWER PLANT: UCI DATASET</h3></center>
    
Features of dataset:
    - T: temperature
    - AP: ambient pressure
    - RH: relative humidity
    - V: vacuum
    - EP: net hourly electrical energy output

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# IMPORT DATASET.
dataset = pd.read_excel('Dataset/Folds5x2_pp.xlsx')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values

#SPLIT THE DATASET.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                   test_size = 0.2,
                                                   random_state = 0)

# BUILD THE NETWROK.
ann = tf.keras.models.Sequential()
# first layer.
first_layer = tf.keras.layers.Dense(units = 6, activation = 'relu')
ann.add(first_layer)
# second layer.
second_layer = tf.keras.layers.Dense(units = 6, activation = 'relu')
ann.add(second_layer)
# outputlayer:
#   FOR BINARY CLASSIFICATION: SIGMOID
#   FOR MULTI-CLASSIFICATION: SOTMAX
#   FOR REGRESSION: NO ACTIVATION FUNCTION
number_output = 1
output_layer = tf.keras.layers.Dense(units = number_output)
# for a regression, the output layer must be only the weighted sum. 
ann.add(output_layer)

# COMPILE THE NETWORK.
ann.compile(optimizer = 'adam', # technique which perform the optimization (such as Gradient Descent).)
            loss = 'mean_squared_error', # in binary classification, the most used is binary_crossentropy.
           )

# TRAIN THE NETWORK.
ann.fit(X_train, 
        y_train, 
        batch_size = 32, # comparison between the true and the redicted value occur before batch_size
       epochs = 100)
```

```python
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
